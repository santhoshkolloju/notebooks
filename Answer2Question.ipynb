```
import json
import pickle 
import glob 
import random 


def create_model_data(squad_file, model_input_file_name,model_output_file_name):
	with open(squad_file, "r") as reader:
		input_data = json.load(reader)
		
	## Data preparation
	input_txtfile = open(model_input_file_name, 'w')
	output_txtfile = open(model_output_file_name, 'w')

	for i in range(len(input_data['data'])):
		for j in range(len(input_data['data'][i]['paragraphs'])):
			context = input_data['data'][i]['paragraphs'][j]['context']
			for k in range(len(input_data['data'][i]['paragraphs'][j]['qas'])):
				question = input_data['data'][i]['paragraphs'][j]['qas'][k]['question']
				answer = input_data['data'][i]['paragraphs'][j]['qas'][k]['answers'][0]['text']
				modelinput_datapoint=context+" ||| "+answer
				modeloutput_datapoint = question 
				input_txtfile.write("%s\n"%modelinput_datapoint.encode("utf8"))
				output_txtfile.write("%s\n"%modeloutput_datapoint.encode("utf8"))
	input_txtfile.close()
	input_txtfile.close()

## 
create_model_data("train-v1.1.json","train_story.txt", "train_summary.txt")
create_model_data("dev-v1.1.json","test_story.txt", "test_summary.txt")
create_model_data("dev-v1.1.json","eval_story.txt", "eval_summary.txt")



		

			
			
			
```
